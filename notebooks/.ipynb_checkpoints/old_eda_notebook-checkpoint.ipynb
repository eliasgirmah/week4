```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
df_train = pd.read_csv('../data/train.csv', low_memory=False)
df_test = pd.read_csv('../data/test.csv', low_memory=False)
df_store = pd.read_csv('../data/store.csv', low_memory=False)

# Display DataFrames' Info
df_train.info()
df_test.info()
df_store.info()

# Summary Statistics
df_train.describe()
df_test.describe()
df_store.describe()

# Missing Values
df_train.isnull().sum()
df_test.isnull().sum()
df_store.isnull().sum()

# Example of EDA - Distribution of Sales
plt.figure(figsize=(10, 6))
sns.histplot(df_train['Sales'], bins=50, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Example of EDA - Sales by Store
plt.figure(figsize=(12, 8))
sns.boxplot(x='Store', y='Sales', data=df_train)
plt.title('Sales by Store')
plt.xlabel('Store')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()

# Example of EDA - Sales over Time
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_train.set_index('Date', inplace=True)
df_train['Sales'].resample('M').mean().plot(figsize=(14, 7))
plt.title('Average Monthly Sales')
plt.xlabel('Date')
plt.ylabel('Average Sales')
plt.show()
```
